```python
import pandas as pd
%pip install pulp
from pulp import *

#Load and inspect the data set

data = pd.read_excel(r"C:\Users\omdre\OneDrive\Documents\GT HOMEWORK\Homework11_ISYE6501\diet.xls")
print(data.head())
```

```python
#Create Foods list
Foods = data['Foods'].tolist()

```

```python
#Inspected Foods list data
print(Foods)
```

```python
#Create food-costs dictinonary
Costs = dict(zip(data['Foods'], data['Price/ Serving']))
```

```python
#Inspected the Costs output
print(Costs)
```

```python
#Create nutrients list
Nutrients = data.columns[3:].tolist()
```

```python
print(Nutrients)
```

```python
#Separate nutrients data
nutrients_data = data.drop(columns=['Foods', 'Price/ Serving', 'Serving Size'])
```

```python
#define min/max for each nutrient

nutrition_limits = {'Calories':{'min': 1500, 'max': 2500},'Cholesterol mg': {'min': 30, 'max': 240},
                    'Total_Fat g':{'min': 20, 'max': 70}, 'Sodium mg': {'min': 800, 'max': 2000}, 
                    'Carbohydrates g':{'min': 130, 'max': 450}, 'Dietary_Fiber g': {'min': 125, 'max': 250}, 
                    'Protein g':{'min': 60, 'max': 100}, 'Vit_A IU': {'min': 1000, 'max': 10000}, 
                    'Vit_C IU':{'min': 400, 'max': 5000}, 'Calcium mg': {'min': 700, 'max': 1500}, 'Iron mg': {'min': 10, 'max': 40}}
```

```python
#Create the problem variable
problem = LpProblem("Diet_Problem", LpMinimize)
```

```python
#Create food variable 
# lowBound=0 means no negative servings, cat='Continuous' allows fractions

food_variables = LpVariable.dicts("Food", Foods, lowBound=0, cat='Continuous')
```

```python
# Create a binary variable to state that food is in the diet
selected_food = LpVariable.dicts('Selected', Foods, cat=LpBinary)
```

```python
#PART 1
#Objective function: minimize the total cost, add it to problem 
#LpSum calculates the total cost of all foods per person
problem += lpSum([Costs[i] * food_variables[i] for i in Foods]),'Total Cost'
```

```python
print(problem)
```

```python
#Constraints:Add nutritional contraints to each nutrient
for nutrient in nutrition_limits.keys():
    if nutrient in Nutrients: 
        min_val = nutrition_limits[nutrient]['min']
        max_val = nutrition_limits[nutrient]['max']
        nutrient_values = data[nutrient].tolist()

        problem += pulp.lpSum([nutrient_values[i] * food_variables[Foods[i]] for i in range(len(Foods))]) >= min_val, f"Min_{nutrient}"
        problem += pulp.lpSum([nutrient_values[i] * food_variables[Foods[i]] for i in range(len(Foods))]) <= max_val, f"Max_{nutrient}"
```

```python
problem.solve()
```

```python
print(f"Status: {LpStatus[problem.status]}")
```

```python
print(f"Total cost: ${value(problem.objective):.2f}")
```

```python
print("\nOptimal Diet (List of Foods):")
for food in Foods:
    if food_variables[food].varValue > 0:
        print(f"  {food}: {food_variables[food].varValue:.2f} servings")
```

```python
#Set Constraints
#If a food is selected, then a minimum of 1/10 serving must be chosen

for food in Foods:
    problem += food_variables[food] <= 1000 * selected_food[food], f"Max_Serving_{food}"
    problem += food_variables[food] >= 0.1 * selected_food[food], f"Min_Serving_{food}"
```

```python
#celery and frozen broccoli
celery_broccoli_group = ['Celery, Raw', 'Frozen Broccoli']
```

```python
problem += lpSum([selected_food[food] for food in celery_broccoli_group]) <= 1
```

```python
#At least 3 kinds of meat/poultry/fish/eggs must be selected.
protein_group = [
    'Roasted Chicken', 'Poached Eggs', 'Scrambled Eggs', 'Bologna,Turkey', 
    'Frankfurter, Beef', 'Ham,Sliced,Extralean', 'Kielbasa,Prk', 'Pork', 
    'Sardines in Oil', 'White Tuna in Water'
]
```

```python
#Set constraint 
problem += lpSum([selected_food[food] for food in protein_group]) >= 3
```

```python
#Solve the new model
problem.solve()
```

```python
status = LpStatus[problem.status]
print(f"\nStatus: {status}")
```

```python
print("\nOptimal Diet (List of Foods):")
for food in Foods:
    if food_variables[food].varValue > 0:
        print(f"  {food}: {food_variables[food].varValue:.2f} servings")
```

```python
print(f"Total cost: ${value(problem.objective):.2f}")
```
